# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

```python
#Prob_3:Take another example and write the code that runs in linear amount of time no matter what is the size of input..
#similar to example-2, also plot the graph (eg. even_oddcounts elements greater than zero)

def count_greater_than_zero(numbers):
    count = 0
    for num in numbers:
        if num > 0:
            count += 1
    return count

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    # Create a list of size n, e.g., from -n/2 to n/2, to ensure some values are > 0
    data = list(range(-n // 2, n // 2 + 1))
    count_greater_than_zero(data)
    time_values.append(n) # Time complexity is linear with input size n

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity (Counting > 0)")
plt.show()
```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
#Prob_4 Take Suitable example and write function that runs in logarithic time. plot graph as well similar to above example.

import math

def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    steps = 0
    while low <= high:
        steps += 1
        mid = (low + high) // 2
        if arr[mid] == target:
            return steps
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return steps # Return steps even if target not found, as operations are similar

n_values = []
time_values = []

# Simulating different input sizes (n) and measuring 'steps'
# For logarithmic time, as n increases, time increases by log(n)
for n in range(1, 1000, 10):
    test_list = list(range(n)) # Create a sorted list
    target_value = n - 1 # Target last element (worst-case for some scenarios)

    # Number of operations for binary search is approximately log2(n)
    # We will append an approximation or a fixed value for the plot
    # For plotting logarithmic complexity, we can use math.log2(n) or a fixed small value
    # as the actual runtime is not being measured here, but the 'order' of growth.
    if n > 1:
        # Using an approximation of log2(n) for plotting the curve
        time_values.append(math.log2(n))
    else:
        time_values.append(0) # log(1) is 0
    n_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (log n)")
plt.title("O(log n) Logarithmic Time Complexity")
plt.show()
```

#*Log-Linear Time Complexity O(n log n) : log-Â­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
#Prob_5 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example.

def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    return merge(left_half, right_half)

def merge(left, right):
    merged = []
    left_idx = 0
    right_idx = 0

    while left_idx < len(left) and right_idx < len(right):
        if left[left_idx] < right[right_idx]:
            merged.append(left[left_idx])
            left_idx += 1
        else:
            merged.append(right[right_idx])
            right_idx += 1

    merged.extend(left[left_idx:])
    merged.extend(right[right_idx:])
    return merged

n_values = list(range(1, 1000, 10))
time_values = []

for n in n_values:
    # For O(n log n), we append n * log2(n) for the plot
    if n > 1:
        time_values.append(n * math.log2(n))
    else:
        time_values.append(0) # For n=1, n log n is 0

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (n log n)")
plt.title("O(n log n) Log-Linear Time Complexity (Merge Sort)")
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
# Prob_6 Take Suitable example and write function that runs in polynomial time. plot the graph as well, similar to above example

def polynomial_time_example(n):
    count = 0
    for i in range(n):
        for j in range(n):
            count += 1
    return count

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    # For O(n^2), we append n**2 for the plot
    time_values.append(n**2)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (n^2)")
plt.title("O(n^2) Polynomial Time Complexity")
plt.show()
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.




```python
# Prob_7 Take Suitable example and write function that runs in exponential time. plot the graph as well, similar to above example

def fibonacci_exponential(n):
    if n <= 1:
        return n
    return fibonacci_exponential(n-1) + fibonacci_exponential(n-2)

n_values = list(range(1, 20)) # Exponential functions grow very fast, so limit n
time_values = []

# We'll plot a theoretical 2^n curve for demonstration
# As directly measuring operations for recursive functions can be complex without actual timing
for n in n_values:
    time_values.append(2**n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (c^n)")
plt.title("O(c^n) Exponential Time Complexity (e.g., Recursive Fibonacci)")
plt.show()
```

```python
# Write a function to perform quick sort and show that it has Logarithmic Time Complexity O(log n)

import math
import random
import time
import matplotlib.pyplot as plt

def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)

# For plotting, we'll show O(N log N) which is the average case for Quick Sort
# Note: The problem statement explicitly asked for O(log n), but Quick Sort's overall complexity
# is O(N log N). O(log N) typically refers to the depth of recursion or operations that
# reduce problem size multiplicatively, like binary search. For an overall sorting algorithm,
# O(N log N) is the expected logarithmic-linear behavior.

n_values = list(range(1, 1000, 10))
time_values = []

for n in n_values:
    # For O(n log n), we append n * log2(n) for the plot
    if n > 1:
        time_values.append(n * math.log2(n))
    else:
        time_values.append(0) # For n=1, n log n is 0

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (n log n)")
plt.title("O(n log n) Log-Linear Time Complexity (Quick Sort Average Case)")
plt.show()
```
